```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import string
```

```python
base_url = "https://www.drugs.com/alpha/{}.html"

urls = []
for letter in string.ascii_lowercase:
    url=base_url.format(letter)
    urls.append(url)

urls
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
```

```python
lis = []
for url in urls:
    response = requests.get(url,headers=headers)
    if response.status_code != 200:
        print("faiiiiiled")
    soup = BeautifulSoup(response.content,'html.parser')
    ul = soup.find('ul' , class_ = 'ddc-list-column-2')
    list = ul.find_all('li')
    lis.extend(list)
    
```

```python

dicts = []
for li in lis:
    li = li.find('a')
    text = li.text
    link = "https://www.drugs.com" + li['href']
    dict =  {'name': text, 'link': link}
    dicts.append(dict)
dicts
```

```python
data_set = pd.DataFrame(dicts)
data_set['uses'] = ''
data_set['inconvinients'] = ''
data_set
```

```python
for index , row in data_set.iterrows():
    url = row['link']
    respo = requests.get(url,headers=headers)
    soup = BeautifulSoup(respo.content,'html.parser')
    cont = soup.find('h2',id='uses')
    if cont:
        use = []
        for sibling in cont.find_next_siblings():
            if sibling.name == 'h2':
                break
            if sibling.name == 'p':
                use.append(sibling.text)
        data_set.at[index,'uses'] = use
    
    war=[]
    cont2 = soup.find('h2',id='side-effects')
    if cont2:
        for sibling in cont2.find_next_siblings():
            if sibling.name =='h2':
                break
            if sibling.name == 'p':
                war.append(sibling.text)
        data_set.at[index,'inconvinients'] = war
```

```python
data_set
```

```python
data_set.to_csv("data_set.csv",index=False)
```
